```python
import numpy as np
import pandas as pd

# read the dataset
df = pd.read_csv("Prayagraj_Traffic_Data.csv")
df.head(15)
```

```python
df.info()
```

```python
# if we have special character like -- , ? in missing value place
# df.replace("?", np.nan, inplace = True)

# but in the data set it is already converted in Nan
```

```python
# this code will check where the missing value is
missing_data = df.isnull()
missing_data.head(25)
```

```python
df.isnull().sum()
```

```python
# count the missing value in each column
for column in missing_data.columns:
    print(f"Column '{column}': {missing_data[column].value_counts()[True]} missing values")
```

```python
# now we have to replace nan value with mean,median & mode

avg_intersection_id = df["intersection_id"].astype('float').mean(axis=0).round(1)

# here axis=0 indicates that the mean should be calculated along the rows (i.e.,
# column-wise)

print("average :", avg_intersection_id)


# convert Nan value by mean
df["intersection_id"].replace(np.nan, avg_intersection_id, inplace = True)
```

```python
# similarly we can replace other column also

avg_latitude = df["latitude"].astype('float').mean(axis=0).round(4)
df["latitude"].replace(np.nan ,avg_latitude , inplace=True)

avg_longitude = df["longitude"].astype('float').mean(axis=0).round(4)
df["longitude"].replace(np.nan ,avg_longitude , inplace=True)

avg_vehicle_count = df["vehicle_count"].astype('float').mean(axis=0).round(0)
df["vehicle_count"].replace(np.nan , avg_vehicle_count ,inplace=True)

avg_average_speed = df["average_speed"].astype('float').mean(axis=0).round(2)
df["average_speed"].replace(np.nan , avg_average_speed ,inplace=True)

avg_signal_cycle_time = df["signal_cycle_time"].astype('float').mean(axis=0).round(1)
df["signal_cycle_time"].replace(np.nan , avg_signal_cycle_time , inplace=True)

avg_green_time = df["green_time"].astype('float').mean(axis=0).round(1)
df["green_time"].replace(np.nan , avg_green_time , inplace=True)

avg_yellow_time = df["yellow_time"].astype('float').mean(axis=0).round(1)
df["yellow_time"].replace(np.nan , avg_yellow_time , inplace=True)

avg_red_time = df["red_time"].astype('float').mean(axis=0).round(1)
df["red_time"].replace(np.nan , avg_red_time , inplace=True)

avg_pollution_level = df["pollution_level"].astype('float').mean(axis=0).round(2)
df["pollution_level"].replace(np.nan , avg_pollution_level , inplace=True)

avg_car_count = df["car_count"].astype('float').mean(axis=0).round(0)
df["car_count"].replace(np.nan , avg_car_count , inplace=True)

avg_bus_count = df["bus_count"].astype('float').mean(axis=0).round(0)
df["bus_count"].replace(np.nan , avg_bus_count , inplace=True)

avg_truck_count = df["truck_count"].astype('float').mean(axis=0).round(0)
df["truck_count"].replace(np.nan , avg_truck_count , inplace=True)

avg_motorcycle_count = df["motorcycle_count"].astype('float').mean(axis=0).round(0)
df["motorcycle_count"].replace(np.nan , avg_motorcycle_count , inplace=True)



df.isnull().sum()
```

```python
# now replace the nan value with most frequent values

df['intersection_name'].value_counts().idxmax()
```

```python
df["intersection_name"].replace(np.nan , "Chowk" , inplace=True)
```

```python
df["weather_condition"].value_counts().idxmax()
```

```python
df["weather_condition"].replace(np.nan, "Clear" , inplace=True)
```

```python
df["day_of_week"].value_counts().idxmax()
```

```python
df["day_of_week"].replace(np.nan,"Thrusday" , inplace=True)
```

```python
df["is_holiday"].value_counts().idxmax()
```

```python
df["is_holiday"].replace(np.nan , 0.0 , inplace=True)
```

```python
# we cannot find timestamp using mean , median & mode

df = df.dropna(subset=['timestamp'])
```

```python
df.info()
```

```python
# Convert the timestamp column to datetime format:
df['timestamp'] = pd.to_datetime(df['timestamp'])

# extract hours from timestamp
df['hour_of_day'] = df['timestamp'].dt.hour


df['is_weekend'] = df['day_of_week'].isin(['Saturday', 'Sunday'])


# Save the cleaned DataFrame to a CSV file
# df.to_csv('cleaned_data.csv')
```

```python
# IQR(Inter Quantile Range)
q1 = df['vehicle_count'].quantile(0.25)
q3 = df['vehicle_count'].quantile(0.75)
iqr = q3-q1

q1,q3,iqr
```

```python
upper_limit = q3 + (1.5*iqr)
lower_limit = q1 - (1.5*iqr)
lower_limit,upper_limit
```

```python
sns.boxplot(df['vehicle_count'])
```

```python
# find the outliers
df.loc[(df['vehicle_count'] > upper_limit) | (df['vehicle_count'] < lower_limit)]
```

```python
# trimming the outlier
new_df = df.loc[(df['vehicle_count'] < upper_limit) & (df['vehicle_count'] > lower_limit)]
print('before removing outlier : ',len(df))
print('after removing outlier : ',len(new_df))
print('outliers : ',len(df) - len(new_df))
```

```python
sns.boxplot(new_df['vehicle_count'])
```

```python
# capping - change the outlier value to upper or lower value
new_df = df.copy()
new_df.loc[(new_df['vehicle_count'] > upper_limit) , 'vehicle_count'] = upper_limit
new_df.loc[(new_df['vehicle_count'] < lower_limit) , 'vehicle_count'] = lower_limit
```

```python
holiday_df = new_df[new_df['is_holiday'] == 1]
event_df = new_df[new_df['event'].isin(['New Year Celebration', 'Local Market Fair'])]
regular_df = new_df[(new_df['is_holiday'] == 0) & (~new_df['event'].isin(['New Year Celebration', 'Local Market Fair']))]
```

```python
holiday_vehicle_count = holiday_df['vehicle_count'].mean()
event_summary = event_df.groupby('event').agg({'vehicle_count': 'mean'}).reset_index()
regular_day_vehicle_count = regular_df['vehicle_count'].mean()
```

```python
comparison_df = pd.DataFrame({
    'Category': ['Holiday', 'Local Market Fair', 'New Year Celebration', 'Regular Day'],
    'Average Vehicle Count': [
        holiday_vehicle_count,
        event_summary.loc[event_summary['event'] == 'Local Market Fair', 'vehicle_count'].values[0],
        event_summary.loc[event_summary['event'] == 'New Year Celebration', 'vehicle_count'].values[0],
        regular_day_vehicle_count
    ]
})
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=comparison_df, x='Category', y='Average Vehicle Count', palette='muted')
plt.title('Average Vehicle Count: Holidays, Special Events, and Regular Days')
plt.xlabel('Category')
plt.ylabel('Average Vehicle Count')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
correlation_pearson = new_df['vehicle_count'].corr(new_df['average_speed'], method='pearson')

plt.figure(figsize=(8, 6))
sns.scatterplot(data=new_df, x='vehicle_count', y='average_speed', color='blue', alpha=0.6)
plt.title(f'Scatter Plot between Vehicle Count and Average Speed\nPearson Correlation: {correlation_pearson:.2f}')
plt.xlabel('Vehicle Count')
plt.ylabel('Average Speed')
plt.grid(True)


plt.show()
```

```python
print(f"Pearson Correlation between vehicle_count and average_speed: {correlation_pearson:.2f}")
```
